# 1. Carga de librerías y configuración general
# ----------------------------------------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Mostrar todas las columnas
pd.set_option('display.max_columns', None)
```

# ----------------------------------------------
# 2. Cargar el dataset
# ----------------------------------------------

```python
df = pd.read_csv("../data/raw/telco.csv")  # o el path correcto 
df.head()
```

# ----------------------------------------------
# 3. Información básica del dataset
# ----------------------------------------------

```python
print("Shape:", df.shape)
print("\nTipos de datos:")
print(df.dtypes)

print("\nResumen estadístico:")
display(df.describe(include='all'))

print("\nValores nulos totales por columna:")
print(df.isnull().sum())

print("\n Valores unicos por columnas categóricas:")
for col in df.select_dtypes(include=['object']).columns:
    print(f"{col}: {df[col].value_counts()} valores únicos")
```

# ----------------------------------------------
# 4. Matriz de calor sobre los datos faltantes
# ----------------------------------------------

```python
plt.figure(figsize=(14, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='Reds')
plt.title('Matriz de calor de valores faltantes', fontsize=16)
plt.show()
```

# ----------------------------------------------
# 5. Tabla con porcentaje de valores faltantes
# ----------------------------------------------

```python
missing_percent = df.isnull().mean() * 100
missing_table = pd.DataFrame({
    'Missing %': missing_percent,
    'Missing Count': df.isnull().sum()
})
missing_table = missing_table[missing_table['Missing Count'] > 0].sort_values(by='Missing %', ascending=False)
display(missing_table)
```
